```python
class Product:
    """상품 정보를 관리하는 클래스"""
    
    def __init__(self, product_id, name, price, weight, stock):

        self.product_id = product_id
        self.name = name
        self.price = price
        self.weight = weight
        self.stock = stock
    
    def get_info(self):
        return {
            "ID": self.product_id,
            "Name": self.name,
            "Price": self.price,
            "Weight": self.weight,
            "Stock": self.stock
        }
        
    
    def update_stock(self, quantity):
        if self.stock < quantity:
            return False
        else:
            self.stock -= quantity
            return True
    
    def apply_discount(self, discount_rate):
        return int(self.price * (1 - discount_rate / 100))
```

```python
class ShoppingCart:
    """장바구니를 관리하는 클래스"""
    
    def __init__(self):
        self.items = []
    
    def add_product(self, product, quantity):
        if product.update_stock(-quantity):
            self.items.append({'product': product, 'quantity': quantity})
            return True
        else:
            print(f"재고가 부족하여 {product.name}을 추가할 수 없습니다.")
            return False
    
    def remove_product(self, product_id):
        for i, item in enumerate(self.items):
            if item['product'].product_id == product_id:
                item['product'].update_stock(item['quantity'])
                return self.items.pop(i)
        return
    
    def get_total_price(self):
        return sum(item['product'].price * item['quantity'] for item in self.items)
    
    def get_total_weight(self):
        return sum(item['product'].weight * item['quantity'] for item in self.items)
    
    def show_cart(self):
        print("\n------ [ 장바구니 목록 ] ------")
        if not self.items:
            print("장바구니가 비어 있습니다.")
            return
        for item in self.items:
            print(f"- {item['product'].name} | 수량: {item['quantity']}개 | 가격: {item['product'].price * item['quantity']:,}원")
        print(f"------------------------------")
```

```python
class Order:
    """주문 정보를 관리하는 클래스"""
    
    def __init__(self, order_id, cart, customer_name):
        self.order_id = order_id
        self.cart = cart
        self.customer_name = customer_name
        self.shipping_fee = 0
    
    def calculate_final_price(self):
        total_price = self.cart.get_total_price()
        total_weight = self.cart.get_total_weight()
        
        self.shipping_fee = 5000 if total_weight > 5.0 else 0
        return total_price + self.shipping_fee
    
    def print_receipt(self):
        final_price = self.calculate_final_price()
        total_product_price = self.cart.get_total_price()
        print(f"\n======= [ 주문 영수증 ] =======")
        print(f"- 주문 ID : {self.order_id}")
        print(f"- 고객명 : {self.customer_name}")
        
        print(f"- 주문 상품 목록 :")
        for item in self.cart.items:
            p = item['product']
            q = item['quantity']
            print(f"> {p.name} | {q}개 | {p.price * q}원")
            
        print(f"------------------------------")
        print(f"- 상품 금액 : {total_product_price:,}원")
        print(f"- 배송비 : {self.shipping_fee:,}원")
        print(f"------------------------------")
        print(f"- 최종 결제 금액 : {final_price:,}원")
        print(f"==============================")
```

```python
# 상품 생성
laptop = Product("P001", "노트북", 1200000, 2.5, 10)
mouse = Product("P002", "무선마우스", 35000, 0.2, 50)
keyboard = Product("P003", "기계식키보드", 89000, 1.0, 30)

# 장바구니 생성 및 상품 추가
cart = ShoppingCart()
cart.add_product(laptop, 1)
cart.add_product(mouse, 2)
cart.add_product(keyboard, 1)

# 장바구니 확인
cart.show_cart()

# 주문 생성
order = Order("ORD20260211001", cart, "홍길동")

# 주문서 출력
order.print_receipt()
```
