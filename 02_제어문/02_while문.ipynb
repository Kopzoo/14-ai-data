```python
# while문 기본 구조
# while 조건문 :
#   수행할문장1
#   수행할문장2
#   수행할문장3
#   ...
# break
```

```python
treehit = 0
while treehit < 10 :
    treehit += 1
    print(f"나무를 {treehit}번 찍었습니다.")
    if treehit == 10 :
        print("나무가 넘어갑니다.")
```

```python
number = 0
prompt = """
1. Add
2. Del
3. List
4. Quit
"""

while number != 4 :
    print(prompt)
    number = int(input("1 ~ 4까지의 숫자 중 하나를 입력해주세요."))
```

```python
# while문 강제로 빠져나가기

#가격 300원
커피 = 5
돈 = 300
while 돈 :
    print("돈을 받았으니 커피를 줍니다.")
    커피 -= 1
    돈 += 300
    print(f"남은 커피는 {커피}개입니다.")
    print(f"현재 돈은 {돈}원입니다.")
    if 커피 == 0 :
        print("판매를 중지합니다.")
        break
```
