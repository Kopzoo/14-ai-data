```python
def greet(name):
    print(f"안녕하세요, {name}님! 반갑습니다.")
    
greet("철수")
```

```python
def calculate(a, b, operator):
    if operator == "+":
        return a + b
    elif operator == "-":
        return a - b
    elif operator == "*":
        return a * b
    elif operator == "/":
        if b == 0:
            return "0으로 나눌 수 없습니다"
        return a / b
    else:
        return "잘못된 연산자입니다"
    
calculate(10, 3, "/")
```

```python
def sum_and_avg(a, b):
    sum = a + b
    avg = sum / 2
    return sum, avg

sum_and_avg(4, 6)
```

```python
def order_coffee(menu, size="Tall", shot=1):
    print(f"[주문] {menu} ({size}, {shot})")

order_coffee("아메리카노")
order_coffee("에스프레소", "short", 3)
```

```python
def get_average(*args):
    if not args:
        return 0
    
    total = sum(args)
    avg = total / len(args)
    return round(avg, 2)

get_average(80, 90, 100)
```

```python
def print_gugudan(dan, start=1, end=9):
    if start > end:
        print("잘못된 범위입니다")
    else:
        for i in range(start, end + 1):
                print(f"{dan} x {i} = {dan * i}")

print_gugudan(7, 9, 2)
```

```python
def left_triangle(height):
    for i in range(1, height + 1):
        print("*" * i)

left_triangle(5)
```

```python
def check_even_odd(num):
    if num % 2 == 0:
        return "짝수"
    else:
        return "홀수"

print(check_even_odd(10))
print(check_even_odd(7))
```

```python
def my_abs(num):
    if num < 0:
        return -num
    else:
        return num

print(my_abs(-3))
```

```python
def find_max(a, b, c):
    if a > b:
        if a > c:
            return a
        else:
            return c
    else:
        if b > c:
            return b
        else:
            return c
        
find_max(3, 8, 5)
find_max(10, 4, 7)
```

```python
def repeat_name(name, count):
    for i in range(1, count + 1):
        print(name)

repeat_name("민수", 3)
```

```python
def print_rectangle(width, height):
    for j in range(1, height + 1):
        print("*" * width)

print_rectangle(4,3)
```

```python
def factorial(n):
    result = 1
    for i in range (1, n + 1):
        result *= i
    return result
    
factorial(0)
```

```python
def list_sum(numbers):
    total = 0
    for n in numbers:
        total += n
    return total

print(list_sum([1, 2, 3, 4, 5]))   # 15
print(list_sum([10, 20, 30]))
```

```python
def get_even_numbers(numbers):
    even = []
    for n in numbers:
        if n % 2 == 0:
            even.append(n)
    return even

print(get_even_numbers([1, 2, 3, 4, 5, 6]))
```

```python
def my_len(text):
    count = 0
    for i in text :
        count += 1
    return count

my_len("python")
```

```python
def count_char(text, ch):
    count = 0
    for char in text:
        if char == ch:
            count += 1
    return count

print(count_char("banana", "a"))  # 3
print(count_char("hello", "l"))
```

```python
def reverse_text(text):
    reverse = ""
    for i in text:
        reverse = i + reverse
    return reverse

print(reverse_text("python"))  # nohtyp
print(reverse_text("abc"))
```
