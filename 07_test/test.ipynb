```python
year = int(input("연도를 입력하세요: "))
if (year % 4  == 0 and year % 100 != 0) or (year % 400 == 0):
    print("윤년")
else:
    print("평년")
```

```python
n = int(input("숫자를 입력하세요: "))

total = 0
if n > 0 :
    for i in range(1, n+1):
        total += i
        if (i<n):
            print(f" {i} +", end = "")
        if (i==n):
            print(f" {i} =", end = " ")
    print(total)
else:
    print("양수만 입력하시오.")
```

```python

def print_gugudan(dan, start=1, end=9):
    if start > end:
        print("잘못된 범위입니다")
    else:
        for i in range(start, end+1):
            print(f"{dan} x {i} = {dan * i}")


print_gugudan(2, 4, 8)
                   
```
