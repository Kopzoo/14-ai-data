```python
students = [
    {"이름" : "김민수", "국어" : 85, "영어" : 90, "수학" : 78},
    {"이름" : "이영희", "국어" : 92, "영어" : 88, "수학" : 95},
    {"이름" : "박철수", "국어" : 78, "영어" : 85, "수학" : 82}
]

total0 = students[0]["국어"] + students[0]["영어"] + students[0]["수학"]
average0 = total0 / 3
total1 = students[1]["국어"] + students[1]["영어"] + students[1]["수학"]
average1 = total1 / 3
total2 = students[2]["국어"] + students[2]["영어"] + students[2]["수학"]
average2 = total2 / 3

print("=== 학생 성적 관리 ===")
print("이름:", students[0]["이름"])
print("국어:", students[0]["국어"], ", 영어:", students[0]["영어"], ", 수학:", students[0]["수학"])
print("총점:", total0, ", 평균:", round(average0, 2))
print("-" * 3)
print("이름:", students[1]["이름"])
print("국어:", students[1]["국어"], ", 영어:", students[1]["영어"], ", 수학:", students[1]["수학"])
print("총점:", total1, ", 평균:", round(average1, 2))
print("-" * 3)
print("이름:", students[2]["이름"])
print("국어:", students[2]["국어"], ", 영어:", students[2]["영어"], ", 수학:", students[2]["수학"])
print("총점:", total2, ", 평균:", round(average2,2))
print("-" * 3)

```

```python
text = "python is great python is fun python is powerful"
words = text.split()
cnt = {}
for word in words:
    cnt[word] = cnt.get(word, 0) + 1
# 해당 부분이 어렵고 복잡해서 아직 안 배웠지만 for문으로 해결하는 방법을 찾아보고 적용했습니다.
print("python: ", cnt["python"], "번", sep="")
print("is: ", cnt["is"], "번", sep="")
print("great: ", cnt["great"], "번", sep="")
print("fun: ", cnt["fun"], "번", sep="")
print("powerful: ", cnt["powerful"], "번", sep="")
```

```python
a_class = {'김철수', '이영희', '박민수', '최지영', '홍길동'}
b_class = {'박민수', '최지영', '정수민', '강호동', '이영희'}

print(a_class & b_class) # 1. 교집합
print(a_class | b_class) # 2. 합집합
print(a_class - b_class) # 3. A 차집합
print(b_class - a_class) # 4. B 차집합
```
