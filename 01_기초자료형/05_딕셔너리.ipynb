```python
# 딕셔너리는 어떻게 만들까?
dic = {"name" : "ken", "phone" : "010-1234-1234", "birth" : "01-01"}
print(type(dic))
```

```python
# 딕셔너리 쌍 추가하기
a = {1: "a"}
a[2] = "b"
print(a)
a["name"] = "ken"
print(a)
a[1] = "c"
print(a)
a[(1,2)] = "hobby"
print(a)
```

```python
# 딕셔너리 요소 삭제하기
print(a)
del a[1]
print(a)
del a["name"]
print(a)
```

```python
# 딕셔너리에서 key를 사용해 value 값 얻기
profile = {
    "name" : "홍길동",
    "age"  : 25,    
}

print(profile["name"])
```

```python
# 딕셔너리에 관련된 함수
# 1. key 리스트 만들기 - keys
profile = {
    "name" : "홍길동",
    "age"  : 25,
    "번호"  : "010-1234-1234"   
}
print(profile.keys())

# 2. value 리스트 만들기 - values
print(profile.values())

# 3. key : value 쌍 모두 얻기 - items
print(profile.items())

# 4. key, value 쌍 모두 지우기 - clear
print(profile.clear())

# 5. key로 value 얻기 - get
profile = {
    "name" : "홍길동",
    "age"  : 25,
    "번호"  : "010-1234-1234"   
}
print(profile.get("name")) # profile["name"]

# 해당 key가 딕셔너리 안에 있는지 조사하기 - in(명령어)
print("name" in profile)

# key로 value 얻기 - pop
temp = profile.pop("name")
print(temp)
print(profile)
```
