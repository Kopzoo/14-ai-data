```python
# 문자열 슬라이싱
a = "Life is too short, you need Python"
print(a[0:4])
print(a[:17])
print(a[19:])
print(a[:])
```

```python
# 문자열 포매팅
print("I eat %d apples." % 3)
print("I eat %s apples." % "five")
number = 3
print("I eat %d apples." % number)
number = 10
day = "three"
print("I ate %d apples, so I was sick for %s days." % (number, day))
```

### 문자열 포맷 코드
- %s : 문자열
- %d : 정수
- %f : 실수

```python
# format 함수
print("I eat {0} apples".format(3))
print("I eat {0} apples".format("five"))
number = 10
day = "three"
print("I ate {0} apples, so I was sick for {1} days".format(number, day))
```

```python
# f-string
name = "홍길동"
age = 30
print(f"나의 이름은 {name}입니다. 나이는 {age}입니다.")
```

```python
# 문자열 함수
a = "hobby"
print(a.count("b"))
print(a.find("b"))
print(a.find("k"))
print(a.index("b"))
print(",".join("abcd"))

a = "hello"
print(a.upper())

a = "HI"
print(a.lower())

a = "   hi   "
print(a.strip())
print(a.lstrip())
print(a.rstrip())

a = "Life is too short"
print(a.replace("Life", "your leg"))
print(a.split())
```
