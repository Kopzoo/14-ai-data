```python
class Calculator:
    def __init__(self):
        self.result = 0

    def add(self, num):
        self.result += num
        return self.result
    
    def sub(self, num):
        self.result -= num
        return self.result
    
    def mul(self, num):
        self.result *= num
        return self.result
    
    def div(self, num):
        self.result /= num
        return self.result

cal1 = Calculator()
cal2 = Calculator()

print(cal1.add(10))
print(cal1.sub(2))
print(cal1.mul(5))
print(cal1.div(4))

print(cal2.add(5))
print(cal2.mul(10))
print(cal2.sub(10))
print(cal2.div(5))
```

```python
class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second

    def add(self):
        result = self.first + self.second
        return result
    
    def sub(self):
        result = self.first - self.second
        return result
    
    def mul(self):
        result = self.first * self.second
        return result
    
    def div(self):
        result = self.first / self.second
        return result
    
```

```python
class FourCal:
   def __init__(self, first, second):
      self.first = first
      self.second = second

   def add(self):
      result = self.first + self.second
      return result


cal4 = FourCal(10, 20)
print(cal4.first) 
print(cal4.second)

cal4.add()
```

```python
class MoreFourCal(FourCal):
    def pow(self):
        result = self.first ** self.second
        return result
```

```python
cal6 = MoreFourCal(10, 2)
print(cal6.add())
print(cal6.pow())
```
